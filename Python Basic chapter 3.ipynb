```python
#debugging 3
current_time_str = input("what is the current time (in hours 0-23)?")
wait_time_str = input("How many hours do you want to wait")

current_time_int = int(current_time_str)
wait_time_int = int(wait_time_str)

final_time_int = current_time_int + wait_time_int

final_answer = final_time_int % 24

print("The time after waiting is: ", final_answer)
```

---

dalam Python sendiri, 1 kesalahan dapat langsung menyebabkan sintaks tidak terpenuhi dan proses terhenti.

## 3.5. Runtime Errors
---
selain SyntaxError, kesalahan lain adalah Runtime Errors yaitu kesalahan yang muncul setelah program dijalankan. kesalahan ini jarang muncul pada program sederhana.

```python
A = int(input("angka pertama"))
B = int(input("angka kedua"))

pembagian = A/B

print("hasil pembagian angka pertama dan kedua", pembagian)
```

terjadi kesalahan karena input angka kedua adalah 0 atau menghasilkan inf

## 3.6. Semantic Errors
---
merupakan kesalahan yang terjadi dimana program akan berjalan namun akan menghasilkan output yang salah atau tidak sesuai dengan perintah tanpa adanya error message.

jenis kesalahan ini lebih sulit diuraikan karena diperlukan pemahaman dan ketelitian

```python
angka = int(input("angka yang akan dibuat persentase"))

persen = angka

print("hasil persentase",persen,"%")
```

terjadi kesalahan hasil dari persentase 12 adalah 0.12 %
