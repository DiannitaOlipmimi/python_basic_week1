## 2.2.1 Order of Operations

---
When more than one operator appears in an expression, the order of evaluation depends on the rules of precedence. Python follows the same precedence rules for its mathematical operators that mathematics does.


```python
print(16 - 2 * 5 // 3 + 1)
```

## 2.4 Function Cells

---

The Python interpreter can compute new values with function calls, using one or more input (arguments).

```python
def square(x):
   return x * x

def sub(x, y):
   return x - y
```

```python
print(square(3))
square(5) # compute only 
print(sub(6, 4)) # function - dg 2 input
print(sub(5, 9))
```

### Complex expression in function

```python
print(square(3) + 2)
print(sub(square(3), square(1+1)))
```

```python
print(square) 
print(square(3)) # () -> invokes function
```
