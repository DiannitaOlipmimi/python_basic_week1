## 2.4 Function Cells

---

The Python interpreter can compute new values with function calls, using one or more input (arguments).

```python
def square(x):
   return x * x

def sub(x, y):
   return x - y
```

```python
print(square(3))
square(5) # compute only 
print(sub(6, 4)) # function - dg 2 input
print(sub(5, 9))
```

```python
print(square(3) + 2)
print(sub(square(3), square(1+1)))
```

## 2.13. Update variabel
---
melakukan update atau perubahan value berdasarkan value yang lama. update variabel dengan menambah value disebut dengan increment; mengurangi disebut decrement. apabila tidak disebutkan banyaknya penambahan atau pengurangan maka nilainya 1. sedangkan bumping variabel adalah menambah 1.

+= adalah increment
-=  adalah decrement

```python
x = 6        
print(x)
x = x + 1    # update x
print(x)
```

```python
x = 6        
print(x)
x += 3       # increment 3 atau x = x + 3
print(x)
x -= 1       # decrement 1
print(x)
```
